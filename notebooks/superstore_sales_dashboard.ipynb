```python
data['Postal Code'] = data['Postal Code'].fillna("Unknown").astype(str) #filling the null values of postal codes with unknown as string type
```

```python
data.isna().sum() #checking null values again/
```

```python
data['Order Date'] = pd.to_datetime(data['Order Date'])
data['Ship Date'] = pd.to_datetime(data['Ship Date'])
```

```python
data['Year'] = data['Order Date'].dt.year
```

```python
data['Month'] = data['Order Date'].dt.month
```

```python
data['Month Name'] = data['Order Date'].dt.month_name()
```

```python
data['Profit'] = data ['Sales']*0.20  ## assuming profit to be 20 percent
```

```python
data['Profit Margin'] = data['Profit']*100 / data['Sales'] ##i think we dont need to add profit margin column since it is considered 20 percent for everything
```

```python
Total_sales = data['Sales'].sum()
Total_sales
```

```python
Total_orders = data['Order ID'].nunique()
Total_orders
```

```python
Total_profit = data['Profit'].sum()
Total_profit
```

```python
Profit_margin = data['Profit Margin']
Profit_margin
```

```python
Avg_order_value = Total_sales / Total_orders
Avg_order_value
```
